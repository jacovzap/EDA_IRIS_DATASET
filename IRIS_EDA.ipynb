# Exploratory Data Analysis

### Objetive
The principal objetive of a EDA is to understand and explore the dataset before applying statistical or machine learning models.

## 1. Inicialization

### 1.1 Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### 1.2 Initial Config

```python
%matplotlib inline

plt.style.use("bmh")
sns.set(style="whitegrid")
```

## 2. Data loading and processing

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.rename({ 'SepalLengthCm': 'sep_l',
            'SepalWidthCm': 'sep_w',
            'PetalLengthCm': 'pet_l',
            'PetalWidthCm': 'pet_w',
            'Species': 'species'
          }, axis = 1, inplace=True)
df.head()
```

## 3. Data Cleaning

```python
df.drop('Id', axis=1, inplace=True)
df.head()
```

## 4. Descriptive Analysis

```python
# plt.figure(figsize=(8.6))
sns.countplot(data=df, x='species', palette='pastel')
# plt.ylabel('Numero de muestras')
plt.title('Numero de muestras por Especie')
plt.xticks(rotation=45)
plt.show
```

### 3.3 Cleaning null variables

```python
df.isnull().sum()
```

## 4. Univariate Analysis

```python
# 1. Histogram with KDE
plt.figure(figsize=(12,6))
plt.subplot(1, 3, 1)
sns.histplot(df['sep_l'], kde=True, bins=20, color='purple')
plt.title('Histogram with KDE')
plt.xlabel('Length of the petal')

# 2. box plot
plt.subplot(1, 3, 2)
sns.boxplot(y=df['sep_l'], color='lightgreen')
plt.title('Boxplot')

# 3. Violin plot
plt.subplot(1, 3, 3)
sns.violinplot(y=df['sep_l'], color='lightblue')
plt.title('Violin plot')

plt.tight_layout()
plt.show()
```

```python
# 1. Histogram with KDE
plt.figure(figsize=(12,6))
plt.subplot(1, 3, 1)
sns.histplot(df['sep_w'], kde=True, bins=20, color='purple')
plt.title('Histogram with KDE')
plt.xlabel('Length of the petal')

# 2. box plot
plt.subplot(1, 3, 2)
sns.boxplot(y=df['sep_w'], color='lightgreen')
plt.title('Boxplot')

# 3. Violin plot
plt.subplot(1, 3, 3)
sns.violinplot(y=df['sep_w'], color='lightblue')
plt.title('Violin plot')

plt.tight_layout()
plt.show()
```
